```python
# operaters
#arithematic operator (+,-,*,/,**,//)
a = 50
b = 40
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
print(a//b)
```

```python
#assigment operators (+=,-=,*=,/=)
a = 30
a += 5
print(a)
a-=20
print(a)
a*=3
print(a)
a/=10
print(a)
```

```python
#comparision operator (==,!=,>,<,>=,<=)
a = 30
b = 40
print(a==b)
print(a!=b)
print(a<b)
print(a>b)
print(a<=b)
print(a>=b)
```
